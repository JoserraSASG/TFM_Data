```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('OVIEDO-POZUELO_excel.xlsx')
print(df)
```

# Process of ETL

```python
team_a = 'Oviedo Div. Oro'
team_b = 'BM Pozuelo Div. Oro'
```

```python
# Remove white spaces at the beginning and end of column names
df.columns = df.columns.str.strip()

# Create a boolean column where each row is True if the team is 'Oviedo Div. Oro' and the event is 'LOST_BALL'
df['TO_A'] = (df['Team'].str.strip() == team_a) & (df['Event'].str.strip().isin(['LOST_BALL', 'ATTACK_FAULT_COMMIT', ' TECHNICAL_MISTAKE_COMMIT']))

# Convert True to 1 and False to 0, then perform a cumulative sum
df['TO_A'] = df['TO_A'].astype(int).cumsum()

# Create a boolean column where each row is True if the team is 'Pozuelo Div. Oro' and the event is 'LOST_BALL'
df['TO_B'] = (df['Team'].str.strip() == team_b) & (df['Event'].str.strip().isin(['LOST_BALL', 'ATTACK_FAULT_COMMIT']))

# Convert True to 1 and False to 0, then perform a cumulative sum
df['TO_B'] = df['TO_B'].astype(int).cumsum()
```

```python
# Filter the DataFrame for only rows where the team is 'Oviedo Div. Oro' and the event is 'ATTACK_FAULT_COMMIT'
attack_fault_o = df[(df['Team'].str.strip() == team_a) & (df['Event'].str.strip() == 'ATTACK_FAULT_COMMIT')]

# Count the number of rows, which is the number of times the team 'Oviedo Div. Oro' has had the event 'ATTACK_FAULT_COMMIT'
count = len(attack_fault_o)

print("The event 'ATTACK_FAULT_COMMIT' has occurred for Oviedo Div. Oro {} times.".format(count))
```

```python
# Create a boolean column where each row is True if the team is 'Oviedo Div. Oro' and the event is 'LOST_BALL'
df['Goal_A'] = (df['Team'].str.strip() == team_a) & (df['Event'].str.strip() == 'GOAL')

# Convert True to 1 and False to 0, then perform a cumulative sum
df['Goal_A'] = df['Goal_A'].astype(int).cumsum()

# Create a boolean column where each row is True if the team is 'Pozuelo Div. Oro' and the event is 'LOST_BALL'
df['Goal_B'] = (df['Team'].str.strip() == team_b) & (df['Event'].str.strip() == 'GOAL')

# Convert True to 1 and False to 0, then perform a cumulative sum
df['Goal_B'] = df['Goal_B'].astype(int).cumsum()
```

```python
# Get a list of all the columns
cols = list(df.columns)

# Move the column "TO_O" to the correct position.
# You should replace index 4 with the index of the column "Event" + 1.
cols.insert(8, cols.pop(cols.index('Goal_A')))
cols.insert(9, cols.pop(cols.index('Goal_B')))

# Now use ix to reorder the columns
df = df.loc[:, cols]
```

```python
# Create a boolean column where each row is True if the team is 'Oviedo Div. Oro' and the event is 'LOST_BALL'
df['GK_A'] = (df['Team'].str.strip() == team_b) & (df['Event'].str.strip() == 'SAVE')

# Convert True to 1 and False to 0, then perform a cumulative sum
df['GK_A'] = df['GK_A'].astype(int).cumsum()

# Create a boolean column where each row is True if the team is 'Pozuelo Div. Oro' and the event is 'LOST_BALL'
df['GK_B'] = (df['Team'].str.strip() == team_a) & (df['Event'].str.strip() == 'SAVE')

# Convert True to 1 and False to 0, then perform a cumulative sum
df['GK_B'] = df['GK_B'].astype(int).cumsum()
print(df)
```

```python
# Get a list of all the columns
cols = list(df.columns)

# Move the column "TO_O" to the correct position.
# You should replace index 4 with the index of the column "Event" + 1.
cols.insert(10, cols.pop(cols.index('GK_A')))
cols.insert(11, cols.pop(cols.index('GK_B')))

# Now use ix to reorder the columns
df = df.loc[:, cols]
```

```python
# Define the events that represent a throw
throws = ["GOAL", "POST_OUT", "SAVE", "SHOT_BLOCKED", "7M_PROVOKE"]

# Create the new column "Lanz_P" initialized with zeros
df['Lanz_B'] = 0

# Initialize the previous possession and the throw flag
previous_possession = df.loc[0, 'Possession']
throw_counted = False

# Loop through the DataFrame
for i in range(1, len(df)):
    # Remove spaces at the beginning and end of the team names and events
    team = df.loc[i, 'Team'].strip()
    event = df.loc[i, 'Event'].strip()
    current_possession = df.loc[i, 'Possession']

    # If the possession has changed, reset the throw flag
    if current_possession != previous_possession:
        throw_counted = False

    # If the event is a throw, the team is "BM Pozuelo Div. Oro", and we haven't counted a throw for this possession
    if event in throws and team == team_b and not throw_counted:
        # Increment the throw counter
        df.loc[i, 'Lanz_B'] = df.loc[i - 1, 'Lanz_B'] + 1
        # Mark that we've counted a throw for this possession
        throw_counted = True
    else:
        # Otherwise, keep the throw counter the same as the previous moment
        df.loc[i, 'Lanz_B'] = df.loc[i - 1, 'Lanz_B']

    # Update the previous possession
    previous_possession = current_possession

# Create the new column "Lanz_P" initialized with zeros
df['Lanz_A'] = 0

# Initialize the previous possession and the throw flag
previous_possession = df.loc[0, 'Possession']
throw_counted = False

# Loop through the DataFrame
for i in range(1, len(df)):
    # Remove spaces at the beginning and end of the team names and events
    team = df.loc[i, 'Team'].strip()
    event = df.loc[i, 'Event'].strip()
    current_possession = df.loc[i, 'Possession']

    # If the possession has changed, reset the throw flag
    if current_possession != previous_possession:
        throw_counted = False

    # If the event is a throw, the team is "Oviedo Div. Oro", and we haven't counted a throw for this possession
    if event in throws and team == team_a and not throw_counted:
        # Increment the throw counter
        df.loc[i, 'Lanz_A'] = df.loc[i - 1, 'Lanz_A'] + 1
        # Mark that we've counted a throw for this possession
        throw_counted = True
    else:
        # Otherwise, keep the throw counter the same as the previous moment
        df.loc[i, 'Lanz_A'] = df.loc[i - 1, 'Lanz_A']

    # Update the previous possession
    previous_possession = current_possession
    print(df)

# Ensure there are no white spaces at the beginning or end of the 'Team' column
df['Team'] = df['Team'].str.strip()

# Filter the events that belong to 'BM Pozuelo Div. Oro'
pozuelo_df = df[df['Team'] == team_b]

# Initialize the attack counter
pozuelo_df['Attacks_B'] = (pozuelo_df['Possession'].shift() != pozuelo_df['Possession']).astype(int).cumsum()

# Now merge the Pozuelo data with the general data
df = pd.merge(df, pozuelo_df[['Attacks_B']], left_index=True, right_index=True, how='left')

# Fill the NA values with the forward fill method that maintains the last valid value
df['Attacks_B'] = df['Attacks_B'].fillna(method='ffill')

# Replace the remaining NaNs (if any) with 0s
df['Attacks_B'] = df['Attacks_B'].fillna(0)
print(df)
```

```python
df['Team'] = df['Team'].str.strip()

# Filter the events that belong to 'Oviedo Div. Oro'
oviedo_df = df[df['Team'] == team_a]

# Initialize the attack counter
oviedo_df['Attacks_A'] = (oviedo_df['Possession'].shift() != oviedo_df['Possession']).astype(int).cumsum()

# Now merge the Pozuelo data with the general data
df = pd.merge(df, oviedo_df[['Attacks_A']], left_index=True, right_index=True, how='left')

# Fill the NA values with the forward fill method that maintains the last valid value
df['Attacks_A'] = df['Attacks_A'].fillna(method='ffill')

# Replace the remaining NaNs (if any) with 0s
df['Attacks_A'] = df['Attacks_A'].fillna(0)
```

```python
print(df)
```

```python
df['%Eff_A'] = df["Goal_A"] / df["Attacks_A"] * 100
df['%Eff_B'] = df["Goal_B"] / df["Attacks_B"] * 100
df['%GK_A'] = df["GK_A"] / df["Attacks_A"] * 100
df['%GK_B'] = df["GK_B"] / df["Attacks_B"] * 100
df['%TO_A'] = df["TO_A"] / df["Attacks_A"] * 100
df['%TO_B'] = df["TO_B"] / df["Attacks_B"] * 100
df['%Lanz_A'] = df["Lanz_A"] / df["Attacks_A"] * 100
df['%Lanz_B'] = df["Lanz_B"] / df["Attacks_B"] * 100
```
